## <u><b>Part 1: Problem Formulation: Customer Churn Prediction Pipeline</b></u>

## 1. Business Problem Definition

Customer churn occurs when an existing customer stops using a company’s services or purchasing its products, ending their relationship with the company. Churn leads to revenue decline, increased cost of acquiring new customers, and potential loss of market share as former customers may influence others and join competitors. Our focus is on addressable churn—where timely intervention can retain these customers, rather than non-addressable churn (e.g., due to death). Predicting and reducing churn is critical for sustainable business growth[2][47].

---

## 2. Key Business Objectives

- Proactively identify customers who are at risk of churning.
- Enable targeted marketing and retention strategies to reduce churn.
- Minimize customer acquisition and retention costs.
- Support strategic decisions with accurate, timely predictions.
- Maintain and grow revenue by improving customer retention[2].

## 4. Expected Outputs from the Pipeline

- **Cleaned Dataset:** Ready for exploratory data analysis (EDA), free of missing or corrupt values.
- **Transformed Feature Set:** Includes engineered features for ML (aggregates, encodings, normalization).
- **Deployable Churn Prediction Model:** A machine learning model that predicts churn (binary classification) for each customer, suitable for integration with business processes and dashboards.

---

## 5. Measurable Evaluation Metrics

- **Accuracy:** Overall proportion of correct predictions.
- **Precision:** Fraction of predicted churns that are actual churns.
- **Recall (Sensitivity):** Fraction of actual churns the model detected.
- **F1 Score:** Harmonic mean of precision and recall (handles imbalanced classes).
- **ROC-AUC:** Quality of risk ranking (probability scores).
- **Churn Rate Before/After Intervention:** Measures model/business impact.
